```python
import matplotlib.pyplot as plt
import numpy as np
```

# Step Function

```python
x = np.arange(-10, 11, 1)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.title('Step Function')
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = np.zeros(len(x))
y[x>=0] = 1

plot_graph(y, "Step (x)")
```

# Sigmoid 

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('X')
plt.ylabel('Sigmoid (X)')
plt.show()
```

# TanH 

```python
def tanh(x):
    return np.tanh(x)

x = np.arange(-10, 10, 0.1)
y = tanh(x)

plt.plot(x, y)
plt.title('Tanh Function')
plt.xlabel('X')
plt.ylabel('TanH (X)')
plt.show()
```

# ReLU 

```python
def relu(x):
    return np.maximum(0,x)

x = np.arange(-10, 10, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('X')
plt.ylabel('ReLU (X)')
plt.show()
```

# ELU 

```python
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-10, 10, 0.1)
y = elu(x)

import matplotlib.pyplot as plt
plt.plot(x, y)
plt.title('ELU Function')
plt.xlabel('X')
plt.ylabel('ELU (X)')
plt.show()
```

# SELU 

```python
ALPHA = 1.67326
SCALE = 1.0507

def selu(x):
    return SCALE * np.where(x > 0, x, ALPHA * (np.exp(x) - 1))

x = np.arange(-10, 10, 0.1)
y = selu(x)

plt.plot(x, y)
plt.title('SELU Function')
plt.xlabel('X')
plt.ylabel('SELU (X)')
plt.show()
```
